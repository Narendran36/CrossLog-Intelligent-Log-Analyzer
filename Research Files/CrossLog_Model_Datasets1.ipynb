```
from urllib.parse import unquote
```

```
def url_parse(query):
    query = unquote(query)
    return query
```

```
def newremove(statement):
  return statement.replace('\n','')
```

```
def whiteremove(statement):
  return statement.replace(' ','')
```

```
df['url'] = df['url'].apply(url_parse)
df['url'] = df['url'].apply(url_parse)
df['url'] = df['url'].apply(newremove)
df['url'] = df['url'].apply(whiteremove)
```

```
def RL(statement):
  statement = str(statement)
  return len(statement)
```

```
def AL(statement):
  statement = str(statement)
  if '?' not in statement:
    return 0
  else:
    loc = statement.find('?')
    return len(statement[loc+1:])
```

```
def NA(statement):
  statement = str(statement)
  if '?' not in statement:
    return 0
  else:
    loc = statement.find('?')
    statement = statement[loc+1:]
    return statement.count('=')
```

```
def PL(statement):
  statement = str(statement)
  if '?' in statement:
    loc = statement.find('?')
    statement = statement[:loc]
    return len(statement)
  else:
    return len(statement)
```

```
def FU(statement):
  statement = str(statement)
  count = 0
  for i in statement:
    if (i.isupper()):
      count += 1
  return count
```

```
def FL(statement):
  statement = str(statement)
  count = 0
  for i in statement:
    if (i.islower()):
      count += 1
  return count
```

```
def FD(statement):
  statement = str(statement)
  count = 0
  for i in statement:
    if (i.isdigit()):
      count += 1
  return count
```

```
def FS(statement):
  statement = str(statement)
  count = 0
  for i in statement:
    if not (i.isalpha() or i.isdigit()):
      count += 1
  return count
```
