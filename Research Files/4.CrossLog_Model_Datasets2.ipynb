```
import pandas as pd
```

```
df = pd.read_csv("drive/MyDrive/stage.csv")
```

```
df = df[['type','url']]
```

```
df['label'] = 0
```

```
df.loc[df['type']=='SQLi','label'] = 1
```

```
from urllib.parse import unquote
```

```
df = df.fillna('NULL')
```

```
def url_parse(query):
    query = unquote(query)
    return query
```

```
def newremove(statement):
  return statement.replace('\n','')
```

```
df['url'] = df['url'].apply(url_parse)
df['url'] = df['url'].apply(url_parse)
df['url'] = df['url'].apply(newremove)
```

```
df['url'] = df['url'].str.lower()
```

```
mylist = list()
```

```
ch = input()
while(ch != '@@@@@'):
    mylist.append(ch)
    ch = input()
```

```
L = set(mylist)
```

```
len(L)
```

```
for l in L:
  df[l] = 0
```

```
def generate(statement,col):
    if col in str(statement):
        return 1
    else:
        return 0
```

```
for col in L:
  df[col] = df['url'].apply(generate, col=col)
```

```
df
```

```
df.to_csv('final_SQLi.csv')
!cp final_SQLi.csv "drive/My Drive/"
```
